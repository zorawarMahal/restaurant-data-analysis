```python
import requests
from bs4 import BeautifulSoup
```

```python
headers = {'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/98.0.4758.82 Safari/537.36'}
response = requests.get("https://www.zomato.com/bangalore/andhra-gunpowder-brigade-road-bangalore/order",headers=headers)
```

```python
content = response.content
soup = BeautifulSoup(content,"html.parser")
```

```python
import os
import json
import pandas as pd
```

```python
data1=[]
data2=[]
data3=[]
data4=[]
data5=[]
data6=[]
data7=[]
data8=[]
```

```python
containers = soup.find_all("div",{"class":"sc-1s0saks-10 cYSFTJ"})
```

```python
from bs4.element import Tag
for container in containers:
  item_name = container.find('h4',attrs={'class':'sc-1s0saks-15 iSmBPS'})
  #print(item_name.text)
  data1.append(item_name.text)

  item_price = container.find('span',attrs={'class':'sc-17hyc2s-1 cCiQWA'})
  #print(item_price.text)
  data2.append(item_price.text)

  item_rating = container.find('div',attrs={'class':'sc-z30xqq-3 bewuUV'})
  #print(item_rating.text)
  if item_rating is not None:
    rating = item_rating.text
  else:
    rating = None
  data3.append(rating)

  noof_rating = container.find('span',attrs={'class':'sc-z30xqq-4 hTgtKb'})
  #print(noof_rating.text)
  if noof_rating is not None:
    noofrating = noof_rating.text
  else:
    noofrating = None
  data4.append(noofrating)

  item_desc = container.find('p',attrs={'class':'sc-1s0saks-12 hcROsL'})
  #print(item_desc.text)
  if item_desc is not None:
    desc = item_desc.text
  else:
    desc = None
  data5.append(desc)

  best_tag = container.find('div',attrs={'class':'sc-2gamf4-0 hILTXF'})
  #print(best_tag.text)
  if best_tag is not None:
    tag = best_tag.text
  else:
    tag = None
  data6.append(tag)
```

```python
for res_name in soup.find_all('h1',attrs={'class':'sc-7kepeu-0 sc-kafWEX kTxZkY'}):
  print(res_name.text)
  data7.append(res_name.text)
```

```python
for res_area in soup.find_all('a',attrs={'href':'https://www.zomato.com/bangalore/brigade-road-restaurants','class':"sc-cmTdod fjhUCy"}):
  print(res_area.text)
  data8.append(res_area.text)
```

```python
q = {'Restaurant Name':data7,'Restaurant Area':data8,'Item name':data1,'Item price':data2,'Rating':data3,'No of Rating':data4,'Item description':data5,'Bestseller tag':data6}
```

```python
df = pd.DataFrame.from_dict(q, orient='index')
df = df.transpose()
```
